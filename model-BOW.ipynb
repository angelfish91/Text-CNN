```python
import sklearn
import joblib as jl
import numpy as np
import pandas as pd
import xgboost as xgb
```

```python
trainset, trainlabel, testset, testlabel = jl.load('./data/data-bow.jl.z') 
```

```python
dtrain = xgb.DMatrix(trainset, trainlabel)
dtest = xgb.DMatrix(testset, testlabel)
```

```python
params={
'booster':'gbtree',
'objective': 'multi:softmax',
'num_class':20, 
'gamma':0.1,  
'max_depth':12, 
'lambda':2,  
'subsample':0.7, 
'colsample_bytree':0.7, 
'min_child_weight':3, 
'silent':0 ,
'eta': 0.2, 
'seed':1000,
'nthread':8,
#'eval_metric': 'error'
}

plst = list(params.items())
num_rounds = 200
watchlist = [(dtrain, 'train'), (dtest, 'val')]

model = xgb.train(plst, dtrain, num_rounds, watchlist, early_stopping_rounds=100)
model.save_model('./model/xgb.model') 

print ("best best_ntree_limit", model.best_ntree_limit) 
```

```python
pred = model.predict(dtest)
print (sklearn.metrics.accuracy_score(pred, testlabel))
```
